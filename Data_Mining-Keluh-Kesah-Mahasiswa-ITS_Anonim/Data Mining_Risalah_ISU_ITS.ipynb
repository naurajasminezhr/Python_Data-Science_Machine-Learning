```python
# Function to categorize issues based on the provided keywords
def categorize_issues(issue):
    # Initialize a dictionary to hold the counts for each category
    issue_categories = {category: 0 for category in kategori_keywords.keys()}

    # Check each category's keywords in the issue
    for category, keywords in kategori_keywords.items():
        for keyword in keywords:
            # Count the occurrence of each keyword in the issue
            if keyword.lower() in issue.lower():
                issue_categories[category] += 1

    # If no keywords are found, categorize under 'Isu-isu Lainnya'
    if all(value == 0 for value in issue_categories.values()):
        issue_categories["Isu-isu Lainnya"] = 1

    return issue_categories

# Apply the categorization function to the combined issues
categorized_data = data['Keluh Kesah Cleaned'].apply(categorize_issues)

# Convert the categorized data into a DataFrame for easier manipulation
categorized_df = pd.DataFrame(list(categorized_data))

# Summarize the count of issues in each category
category_counts = categorized_df.sum().sort_values(ascending=False)
category_counts
```

```python
normalized_word = pd.read_csv("normalisasi2.csv", encoding = 'latin1')

normalized_word_dict = {}
for index, row in normalized_word.iterrows():
  if row[0] not in normalized_word_dict:
    normalized_word_dict[row[0]] = row[1]

def normalized_term(document):
  return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

data['normalisasi'] = data['Tokens'].apply(normalized_term)
data
```

```python
#remove stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopword = stopwords.words('indonesian')
txt_stopword = pd.read_csv("stopword_risalah.txt", names = ["stopwords"], header = None)

stopword.extend(["nya", "oke", "ada", "nih", "ni", "kenapa", "wkwk", "pagi", "alasan", "biar", "eh"])

stopword.extend(txt_stopword["stopwords"][0].split(' '))
stopword = set(stopword)

def stopwords(text):
  text = [word for word in text if word not in stopword]
  return text
data['Stopwords'] = data['normalisasi'].apply(stopwords)
data
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import pandas as pd

# Membuat stemmer dari Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming pada satu dokumen
def stem_document(doc):
    stemmed_words = [stemmer.stem(term) for term in doc]
    return stemmed_words

# Misalkan df adalah DataFrame Anda
# Asumsi bahwa kolom 'Stopwords' berisi list kata-kata
data['Stemmed'] = data['Stopwords'].apply(stem_document)

# Menampilkan hasil stemming
print(data.head())
```
