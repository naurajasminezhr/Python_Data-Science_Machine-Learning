```python
# 1.1.6 Mengenal Perulangan while
# Coba Ketik
# Catatan: "//" Adalah Floor Division
#
x = 256
total = 0
while x > 0:
    if total > 500:
        break
    total += x
    x = x // 2
    print(total)
    print(x)
```

```python
# 1.2.3 Dictionary
# List adalah struktur data yang dibuat dengan "Key" dan "Value"
# Biasa dikenal dengan HashMap / Associative Array di bahasa pemograman lain
# Cara membuat menggunakan {"Key": "Value"}
#
# Coba ketik
#
karyawan = {"nama": "Ical", "asal": "Surabaya"}
print(karyawan)
```

* API C untuk menghubungkan NumPy dengan pustaka yang ditulis dalam C, C++, atau FORTRAN.

```python
tips = pd.read_csv('https://raw.githubusercontent.com/WilliamQLiu/python-examples/master/pandas/tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
party_counts = party_counts.loc[:, 2:5]
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```python
party_pcts.plot.bar()
```

```python
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```python
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
```

```python
sns.set(style="whitegrid")
tips['tip_pct'].plot.hist(bins=50)
```

```python
tips['tip_pct'].plot.density()
```

```python
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))
sns.distplot(values, bins=100, color='k')
```

```python
sns.distplot(comp2, bins=100, color='k')
```

```python
macro = pd.read_csv('https://raw.githubusercontent.com/BrambleXu/pydata-notebook/master/examples/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

```python
sns.regplot('m1', 'unemp', data=trans_data)
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

```python
sns.factorplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.factorplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.factorplot(x='tip_pct', y='day', kind='box', data=tips[tips.tip_pct < 0.5])
```
