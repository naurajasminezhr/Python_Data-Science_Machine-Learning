# **2. SQL Primer**

## **2.1 Basic Query**

### DISTINCT Clause

```python
# Menggunakan "Select Distinct"
# Mendapatkan unik department_id dari tabel employees
uniq_dept = pd.read_sql_query('SELECT DISTINCT department_id FROM employees', conn)
uniq_dept
```

### WHERE Clause

```python
# Menggunakan Where Clause
# memfilter data employees yang memiliki salary diatas 10000
above_10000 = pd.read_sql_query('''SELECT * 
                                    FROM employees 
                                    WHERE salary > 10000
                                ''', conn)
above_10000
```

```python
# Menggunakan kombinasi WHERE dan AND
# Memfilter data employees yang memiliki salary di atas 10000 dan berasal dari departemen 90 (executive)
executive_and_10000 = pd.read_sql_query('''SELECT * 
                                        FROM employees
                                        WHERE salary > 10000 AND department_id = 90
                                    ''', conn)
executive_and_10000
```

```python
# Menggunakan kombinasi WHERE dan OR
# Memfilter data employees yang memiliki salary di atas 10000 atau berasal dari departemen 80 (sales)
sales_or_10000 = pd.read_sql_query('''SELECT * 
                                        FROM employees
                                        WHERE salary > 10000 OR department_id = 80
                                    ''', conn)
sales_or_10000
```

```python
# Menggunakan kombinasi WHERE dan IN
# Memfilter data employees yang memiliki ID 100,150,200
emp = pd.read_sql_query('''SELECT * 
                            FROM employees
                            WHERE employee_id IN (100,150,200)
                        ''', conn)
emp
```

```python
# Menggunakan kombinasi WHERE dan IN
# Memfilter data employees yang memiliki ID 100,150,200
emp = pd.read_sql_query('''SELECT * 
                            FROM employees
                            WHERE employee_id BETWEEN 100 AND 200
                        ''', conn)
emp
```

```python
# Menggunakan kombinasi WHERE BETWEEN AND
# Memfilter data employees yang memiliki salary diantara range 5000 hingg 10000
emp_5000_10000 = pd.read_sql_query('''SELECT * 
                                      FROM employees
                                      WHERE salary BETWEEN 5000 AND 10000
                                  ''', conn)
emp_5000_10000
```

### GROUP BY Clause

```python
# Mengggunakan GROUP BY clause
# Mendapatkan rata-rata salary masing-masing department
dept_avg = pd.read_sql_query('''SELECT 
                                  department_id,
                                  AVG(salary)
                                FROM employees
                                GROUP BY department_id
                            ''', conn)
dept_avg
```

### HAVING Clause

```python
# Mengggunakan HAVING clause
# Mendapatkan department apa saja yang memiliki rata-rata salary minimum 10000
dept_avg_min10000 = pd.read_sql_query('''SELECT 
                                          department_id,
                                          AVG(salary)
                                        FROM employees
                                        GROUP BY department_id
                                        HAVING AVG(salary) >= 10000
                                    ''', conn)
dept_avg_min10000
```

### ORDER BY Clause

```python
# Mengggunakan GROUP BY clause
# Mendapatkan rata-rata salary masing-masing department dan mengurutkannya dari salary yang paling tinggi ke rendah
dept_avg_order = pd.read_sql_query('''SELECT 
                                        department_id,
                                        AVG(salary)
                                      FROM employees
                                      GROUP BY 1
                                      ORDER BY 2 DESC
                                  ''', conn)
dept_avg_order
```

### LIKE 

```python
# Mengggunakan LIKE
# Mendapatkan data karyawan dengan job_id "...IT..."
job_it = pd.read_sql_query('''select 
                                *
                              from employees
                              WHERE job_id LIKE '%it%'
                          ''', conn)
job_it
```

```python
emp_komisi = pd.read_sql_query('''SELECT
                                    employee_id,
                                    first_name AS "First Name",
                                    last_name,
                                    commission_pct,
                                    CASE
                                      WHEN commission_pct >= 0.40 THEN "high"
                                      WHEN commission_pct >= 0.20 THEN "medium"
                                      ELSE "low" 
                                      END
                                      AS commission_class
                                  FROM employees
                                  WHERE commission_class = "medium"
                              ''', conn)
emp_komisi
```

### LEFT JOIN

```python
# Menggunakan LEFT JOIN
# Mendapatkan unik department_id dari tabel employees serta mendapatkan nama dapartemen dari tabel departments
# uniq_dept_name = pd.read_sql_query('''SELECT 
#                                         DISTINCT emp.department_id,
#                                         dept.depart_name AS department_name
#                                       FROM employees emp
#                                       LEFT JOIN departments dept ON dept.department_id=emp.department_id
#                                       ORDER BY 1 ASC
#                                     ''', conn)
# uniq_dept_name
```

```python
uniq_dept_name = pd.read_sql_query('''SELECT
                                        DISTINCT emp.department_id,
                                        dept.depart_name AS "Department Name"
                                      FROM employees emp
                                      LEFT JOIN departments dept ON dept.department_id=emp.department_id
                                      ORDER BY 1
                                  ''', conn)
uniq_dept_name
```

### INNER JOIN

```python
# Menggunakan INNER JOIN
# Mendapatkan unik department_id dari tabel employees tetapi department_id harus terdaftar di tabel depatments 
# serta mendapatkan nama dapartemen dari tabel departments
uniq_dept_name = pd.read_sql_query('''SELECT 
                                        distinct emp.department_id,
                                        dept.depart_name AS department_name
                                      FROM employees emp
                                      INNER JOIN departments dept 
                                        ON dept.department_id=emp.department_id
                                      ORDER BY 1 ASC
                                    ''', conn)
uniq_dept_name
```

### UNION

```python
# Menggunakan UNION
# 1. Mendapatkan unik department_id dari tabel employees serta mendapatkan nama dapartemen dari tabel departments
# 2. Mendapatkan unik department_id dan nama department dari tabel departments
# UNION 1 dan 2
dept_name = pd.read_sql_query('''SELECT 
                                    DISTINCT emp.department_id,
                                    dept.depart_name AS department_name
                                  FROM employees emp
                                  LEFT JOIN departments dept ON dept.department_id=emp.department_id

                                  UNION

                                  SELECT 
                                    DISTINCT department_id,
                                    depart_name AS department_name
                                  FROM departments
                              ''', conn)
dept_name
```

```python
# Menggunakan UNION
# 1. Mendapatkan unik department_id dari tabel employees serta mendapatkan nama dapartemen dari tabel departments
# 2. Mendapatkan unik department_id dan nama department dari tabel departments
# UNION 1 dan 2
dept_name = pd.read_sql_query('''SELECT 
                                    DISTINCT emp.department_id,
                                    dept.depart_name AS department_name
                                  FROM employees emp
                                  LEFT JOIN departments dept ON dept.department_id=emp.department_id

                                  UNION

                                  SELECT 
                                    DISTINCT department_id,
                                    manager_id
                                  FROM departments
                              ''', conn)
dept_name
```

### SUBQUERY

```python
# Menggunakan SUBQUERY
# 1. Mendapatkan rata-rata salary masing-masing department
# 2. Melabelkan salary setiap employee yang di atas rata-rata salary department sebagai "above average" dan di bawah rata-rata sebagai "under average"
# 3. Menghitung berapa employee yag "above average" dan "under average"
jumlah = pd.read_sql_query('''SELECT
                                list.salary_class,
                                COUNT(DISTINCT list.employee_id) AS jumlah_pegawai
                              FROM
                                  (SELECT
                                      emp.employee_id,
                                      emp.first_name,
                                      emp.last_name,
                                      emp.salary,
                                      ds.average_salary,
                                      CASE 
                                          WHEN emp.salary >= ds.average_salary THEN "above average"
                                          ELSE "under average"
                                          END AS salary_class
                                  FROM employees emp
                                  LEFT JOIN 
                                      (SELECT 
                                          department_id,
                                          AVG(salary) AS average_salary
                                        FROM employees
                                        GROUP BY department_id) ds 
                                  ON ds.department_id=emp.department_id) list
                              GROUP BY 1
                          ''', conn)
jumlah
```

```python
jumlah = pd.read_sql_query('''SELECT
                                list.salary_class,
                                COUNT(DISTINCT list.employee_id)
                              FROM (SELECT
                                      emp.employee_id,
                                      emp.first_name,
                                      emp.last_name,
                                      emp.department_id,
                                      emp.salary,
                                      ds.average_salary,
                                      CASE 
                                          WHEN emp.salary >= ds.average_salary THEN "above average"
                                          ELSE "under average"
                                          END AS salary_class
                                  FROM employees emp
                                  LEFT JOIN (SELECT
                                                department_id,
                                                AVG(salary) AS average_salary
                                              FROM employees
                                              GROUP BY 1) ds
                                  ON ds.department_id = emp.department_id) list
                                GROUP BY 1
                          ''', conn)
jumlah
```
