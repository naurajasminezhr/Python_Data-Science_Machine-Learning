**CASE FOLDING AND CLEANING TEXT**

```python
import pandas as pd
import re

# Define a cleaning function
def clean_text(text):
    # Remove usernames
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)

    # Remove URLs (http and https)
    text = re.sub(r'http\S+|https\S+', '', text)

    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Replace HTML character codes with quotes
    text = re.sub(r'&quot;', '"', text)
    text = re.sub(r'&amp;', '&', text)

    # Replace newlines with spaces
    text = text.replace('\n', ' ')

    # Convert text to lowercase
    text = text.lower()

    # Remove single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)

    # Split and rejoin words to remove extra spaces
    text = ' '.join(text.split())

    return text

# Specify the path to your CSV file
file_path = "tweets-data/cocoki.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=";")

# Apply the cleaning function to the 'full_text' column and create a new 'cleaned_text' column
df['cleaned_text'] = df['full_text'].apply(clean_text)

# Display the DataFrame with the cleaned text column
display(df[['full_text', 'cleaned_text']])
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
# Tokenize the 'cleaned_text' column
df['tokens'] = df['cleaned_text'].apply(word_tokenize)

# Display the DataFrame with the tokens
display(df[['full_text', 'tokens']])
```

**Normalisasi**

```python
normalized_word = pd.read_csv("normalisasi.csv", encoding = 'latin1')

normalized_word_dict = {}
for index, row in normalized_word.iterrows():
  if row[0] not in normalized_word_dict:
    normalized_word_dict[row[0]] = row[1]

def normalized_term(document):
  return [normalized_word_dict[term] if term in normalized_word_dict else term for term in document]

df['normalisasi'] = df['tokens'].apply(normalized_term)
df
```

**Remove Stopwords**

```python
#remove stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopword = stopwords.words('indonesian')
txt_stopword = pd.read_csv("stopword.txt", names = ["stopwords"], header = None)

stopword.extend(["nya", "oke", "ada", "nih", "ni", "kenapa", "wkwk", "pagi", "alasan", "biar", "eh"])

stopword.extend(txt_stopword["stopwords"][0].split(' '))
stopword = set(stopword)

def stopwords(text):
  text = [word for word in text if word not in stopword]
  return text
df['Stopwords'] = df['normalisasi'].apply(stopwords)
df
```
